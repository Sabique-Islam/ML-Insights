#Precision-Recall Curve
## ROC AUC (Area under the ROC curve)
- Shows the performance of a classification model at all classification thresholds

- The curve is plotted by two parameters:
-- True Positive Rate = TP/(TP+FN) [TP: true positive, FN: false negative]
-- False Positive Rate = FP/(FP+TN) [FP: false positive, TN: true negative]


- Best case: TPR: 1, FPR: 0

### ROC, AUC - Limitations

1. **Poor Performance with Imbalanced Data**
- When the dataset is highly imbalanced (ex: 95% negative and 5% positive), the ROC curve can be overly optimistic. Even a model that predicts mostly negatives can achieve a high AUC score, but it might not perform well in real-world scenarios.

2. **Doesn't Differentiate Between Practical Costs of False Positives and False Negatives**
- ROC AUC gives equal importance to False Positives (FP) and False Negatives (FN), but in many applications, they have different costs.
- Example: In medical diagnosis, a False Negative (missed disease) is more severe than a False Positive (false alarm).
PR AUC (Precision-Recall AUC) is often better in such cases.

3. **Can Be Misleading When Decision Thresholds Matter**
- The ROC curve is threshold-independent, meaning it evaluates performance over all possible decision thresholds.
- However, in real applications, we often use a fixed threshold (e.g., 0.5 for classification). The PR curve is more informative when focusing on how precision and recall trade-off at different thresholds.

4. **Not Always Useful for Rare Event Detection**
- In fraud detection, cybersecurity, or medical applications where positive cases are rare, ROC AUC can be misleading.
- **The Precision-Recall (PR) curve is preferred because it focuses on the minority class.**

5. **Doesn't Capture Model Confidence**
- Two models can have the same ROC AUC but different probability distributions. One model might assign higher confidence to correct predictions, but ROC AUC wouldn't reflect that difference.


## Precision-Recall

- Recall/True Positive Rate (x) = TP/(TP+FN)
- Precision (y): TP/(TP+FP) (replacement for FPR)
-- Precision is how many instances were correctly identified out of everything predicted.
-- Perfect Classifier: x=1, y=1, AUC=1 (max)

# Gradient Flow Plot

- Its a visualisation tool used in deep learning to analyse the behavior of gradients during the training of a neural network. Gradients represent direction and mangnitude of the steepest increase in the loss function wrt the model's parameters (eg: weights and biases).
- In training, gradients are computed thru backpropagation and used by optimisers to update parameters and minimise loss

### Gradient flow visualises:
- The magnitude of gradients at different layers or parameters.
- How gradients change over time.
- Potential issues like vanishing or exploding gradients, which can hinder training.

**This plot is crucial for diagnosing training stability, understanding model learning dynamics, and identifying architectural or optimization issues in deep neural networks.**
